# 12 What is the price difference between desi wheat and kalyan HYV state wise?


```python
comp=df[(df["Commodity"]=="Wheat")]
```

```python
comp
```

```python
st=comp["State"].unique()
no=[]

for s in st:
    val=comp[(comp["State"]==s)]["Variety"].nunique()
    if val==1:
        no.append(s)
```

```python
no
```

```python
ind1=comp[(comp["State"]=="Manipur")].index
```

```python
ind1
```

```python

comp=comp.drop(ind1,axis=0)
```

```python
ind2=comp[(comp["State"]=="Meghalaya")].index
```

```python
ind2
```

```python
comp=comp.drop(ind2,axis=0)
```

```python
comp
```

```python
comp['Year'] = pd.DatetimeIndex(comp['Date']).year
```

```python
comp
```

```python
state=comp["State"].unique()
date=comp["Year"].unique()
```

```python
sta=[]
price=[]
dat=[]
for s in state:
    for d in date:
        
        a=comp[(comp["Commodity"]=="Wheat")&(comp["State"]==s)&(comp["Variety"]=="Desi")&(comp["Year"]==d)]["Retail Price"].mean()
        b=comp[(comp["Commodity"]=="Wheat")&(comp["State"]==s)&(comp["Variety"]=="Kalyan HYV")&(comp["Year"]==d)]["Retail Price"].mean()
        diff=a-b
        if diff>=0:
            print("Retail price difference of wheat variety desi and kalyan in %s at %d was %0.3f"%(s,d,diff))
            sta.append(s)
            price.append(diff)
            dat.append(d)
        else:
            print("Retail price difference of wheat variety desi and kalyan in %s at %d was %0.3f"%(s,d,abs(diff)))
            sta.append(s)
            price.append(abs(diff))
            dat.append(d)
```

```python

data = {"State":sta,"Year":dat,"Price Difference":price}
```

```python

price_diff=pd.DataFrame(data)
```

```python

price_diff
```

```python
import plotly.express as px
  

fig = px.bar(price_diff, x="Year", y="Price Difference", 
             color="State", barmode = 'group')
  
fig.show()
```

# 13 Highest and lowest price difference year wise

```python

one=price_diff["Year"].unique()
```

```python
for d in one:
    max=price_diff[(price_diff["Year"]==d)]["Price Difference"].max()
    state=price_diff[(price_diff["Year"]==d)&(price_diff["Price Difference"]==max)]["State"].unique()
    min=price_diff[(price_diff["Year"]==d)]["Price Difference"].min()
    state1=price_diff[(price_diff["Year"]==d)&(price_diff["Price Difference"]==min)]["State"].unique()
    
    
    print("Highest price difference in %s was::%0.3f state::%s"%(d,max,state))
    print("Lowest price difference in %s was::%0.3f state::%s"%(d,min,state1))
```

# 14 How many variety of coffee is available and what is their price differences(Year Wise )?

```python
df
```

```python
df[(df["Commodity"]=="Coffee")]["Variety"].unique()
```

```python
#year wise  price difference
df9=df[(df["Commodity"]=="Coffee")]
df9
```

```python
df9['Date'] = pd.DatetimeIndex(df9['Date']).year
```

```python
df9
```

```python
date=df9["Date"].unique()
```

```python
year=[]
price=[]
for d in date:
    a=df9[(df9["Commodity"]=="Coffee")&(df9["Variety"]=="Instant")&(df9["Date"]==d)]["Retail Price"].mean()
    b=df9[(df9["Commodity"]=="Coffee")&(df9["Variety"]=="Powdered")&(df9["Date"]==d)]["Retail Price"].mean()
    diff=a-b
    if diff>=0:
        #diff=("%0.2f"%(diff))
        
        print("Retail price difference of coffee variety Instant and Powdered in %d was %f"%(d,diff))
        year.append(d)
        price.append(diff)
        
    else:
        print("Retail price difference of coffee variety Instant and Powdered in %d was %0.3f"%(d,abs(diff)))
        year.append(d)
        price.append(abs(diff))
```

```python
pr = []

for i in range(len(price)):
    p = float(("%0.2f"%(price[i])))
    pr.append(p)
```

```python
data1 = {"Year":year,"Price":pr}
```

```python
price_dif=pd.DataFrame(data1)
```

```python
price_dif
```

```python
plt.figure(figsize=(10,15))
plt.bar("Year","Price",data=price_dif)
plt.xlabel("Year")
plt.ylabel("Price difference")
plt.title("Retail Prices difference of coffee variety instant and powderd in year wise")
plt.show()
```

# 15 Which state has lowest price of Bajra according to yearwise?

```python
df
```

```python
df11=df[(df["Commodity"]=="Bajra")]
```

```python
df11
```

```python
dat=df11["Date"].unique()
```

```python
dat
```

```python
pri=[]
stat=[]
year=[]
for d in dat:
    price=df11[(df11["Date"]==d)]["Retail Price"].min()
    sta=df11[(df11["Date"]==d)&(df11["Commodity"]=="Bajra")&(df11["Retail Price"]==price)]["State"].unique()
    
    
    print("Minimum price of bajra in %d:: state::%s price:: %0.3f"%(d,sta,price))
    pri.append(price)
    stat.append(sta)
    year.append(d)
```

```python
state = []
for i in stat:
    state.append(i[0])
```

```python
#to plot multiple state with same value for bajra 

state[5] = "Bihar, Orissa, Karnataka, Rajathan"
state[11] = "Rajasthan, Uttar Pradesh"
state[12] = "Panjab Rajshthan"
state[13] = "Panjab Rajshthan"
state[15] = "Haryana, Rajasthan"
```

```python
plt.figure(figsize = (12,7))
bar_plot =  plt.bar(year,pri)


def autolabel(rects):
    for idx,rect in enumerate(bar_plot):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                state[idx],
                ha='center', va='bottom', rotation=90)

autolabel(bar_plot)


plt.show()
```
